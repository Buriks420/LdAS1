```python
# Input: Minta pengguna memasukkan jumlah kelereng Aldi.
# Hitung:
# Hitung jumlah kelereng Budi.
# Hitung jumlah kelereng Anto.
# Hitung jumlah kelereng Agung.
# Output: Tampilkan hasil perhitungan jumlah kelereng Budi, Anto, dan Agung.
Aldi = int(input("Masukkan jumlah kelereng Aldi: "))
Budi = Aldi - 15
Anto = (Budi + Aldi) * 2
Agung = Aldi + Budi + Anto - 5

print(Aldi, Budi, Anto, Agung)
```
