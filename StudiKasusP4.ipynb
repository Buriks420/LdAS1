```python
def Variabel(A, B):
    if A > B:
        C = A * B
    else:
        C = A + B
    D = C * C
    return C, D

print("Hasil dari Variabel a. 4 dan 6 adalah : ", Variabel(4, 6))
print("Hasil dari Variabel b. 3 dan 5 adalah : ", Variabel(3, 5))
print("Hasil dari Variabel c. 9 dan 6 adalah : ", Variabel(9, 6))
```

```python
def Variabel(A, B):
    if A + B < 10:
        C = A - B
    else:
        C = A + B
    D = 2 * C + B
    return C, D

print("Hasil dari Variabel a. 2 dan 4 adalah : ", Variabel(2, 4))
print("Hasil dari Variabel b. 5 dan 7 adalah : ", Variabel(5, 7))
print("Hasil dari Variabel c. 8 dan 5 adalah : ", Variabel(8, 5))
```

```python
lama = int(input("Mau Berapa Jam kamu untuk menyewa PC disini? "))
tarif1 = 6000
tarif2 = 5000
if lama <= 3:
    total = (lama * tarif1)
elif lama >= 3:
    total = (lama - 3) * tarif2 + 18000

print(f"Bayar : Rp.{total}")

```

```python
gajipokok = 5000000
produk = int(input("Masukan banyak produk yang terjual : "))
harga = int(input("Masukan Harga satuan produk :"))

if produk >= 100:
    bonus = (produk * harga) * 0.2
else:
    bonus = (produk * harga) * 0.1

gaji = gajipokok + bonus
print(f"Gaji kamu adalah {gaji:.0f}")
```

```python
GajiPokok = int(input("Masukan Gaji Pokok"))
JamKerja = int(input("Masukan Total Jam Kerja"))
Lembur = 20000

if JamKerja >= 200:
    JamKerja = ((JamKerja - 200)) * Lembur
else:
    JamKerja = 0
    
Gaji = GajiPokok * 0.2
GajiSebelumPajak = Gaji + GajiPokok 
Pajak = GajiSebelumPajak * 0.1
GajiSetelahPajak = GajiSebelumPajak - Pajak
TotalGaji = GajiSetelahPajak + JamKerja
print(f"Gaji Kamu adalah : {(GajiSebelumPajak):.0f} dikurangi pajak 10% jadi = {GajiSetelahPajak:.0f} ")
print(f"Gaji Lembur Kamu adalah : {JamKerja}")
print(f"Total Gaji kamu adalah : {TotalGaji:.0f}")
```
