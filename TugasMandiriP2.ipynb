```python
brt = int(input("Masukan Berat Telur : "))
hrg = int(input("Masukan Harga Telur : "))
ongkos = int(input("Masukan Biaya Transportasi : "))
uang = int(input("Berapa banyak Uang Ibu?"))

telor = hrg * brt
angkot = telor + (ongkos * 2 )
sisa = uang - angkot

print(f"Sisa uang Ibu adalah : Rp.{sisa}") 
```
