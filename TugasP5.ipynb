```python
while True:
    tinggi = int(input("Masukan tinggi segitiga"))
    if 1 <= tinggi <= 100:
        break
    else:
        print("Tinggi segitiga harus antara 1 dan 100.")
segitiga = input("Masukan karakter khusus untuk membuat segitiga ")
for i in range(tinggi):
    spasi = ' ' * (tinggi - i - 1)
    karakter = segitiga * (2 * i + 1)
    print(spasi + karakter)
```
